#### Converting into date format

```python
def convert_to_dateform(df,column):
   df[column] = pd.to_datetime(df[column] ,errors='coerce')
```

```python
dates=['Expiration_date','Date']

for i in dates:
   convert_to_dateform(data,i)
```
